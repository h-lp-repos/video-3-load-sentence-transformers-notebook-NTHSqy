# Load Sentence-Transformers model and run single-batch inference

```python
from sentence_transformers import SentenceTransformer
import numpy as np
```

```python
model_name = "sentence-transformers/all-MiniLM-L6-v2"
print(f"Loading model {model_name}")
model = SentenceTransformer(model_name)
print("Model loaded")
```

```python
batch_texts = [
    "Este es el primer documento.",
    "Este es el segundo documento."
]
```

```python
embeddings = model.encode(batch_texts, batch_size=32, show_progress_bar=True)
print(f"Embeddings shape: {embeddings.shape}")
print("First embedding values:", embeddings[0][:6])
```

```python
import os
np.savez_compressed("embeddings.npz", embeddings=embeddings)
print("Saved embeddings to embeddings.npz")
os.system("ls -lh embeddings.npz")
```
